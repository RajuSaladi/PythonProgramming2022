## 1. Write a Program to calculate the sum of items of the list [20, 50, 100, -10]

```python
sum_of_list = 0
input_list = [20, 50, 100, -10]
for item in input_list:
  print("Item value ", item)
  print("Before:", sum_of_list)
  sum_of_list = sum_of_list + item
  print("After:", sum_of_list)

print("Sum of list is ", sum_of_list)
```

## 2. write a program to count the number of even and odd numbers in the list [1, 5, 2, 9, 16, 21, 50]
  Hint: Odd numbers give remainder 1 when divided by 2
            Even numbers give remainder 0 when divided by 2

```python
input_list =  [1, 5, 2, 9, 16, 21, 50]
even_num_count = 0
odd_num_count = 0

for item in input_list:
  if item%2 == 0:
    even_num_count = even_num_count + 1
  elif item%2 == 1:
    odd_num_count = odd_num_count + 1

print("Even number count", even_num_count)
print("Odd number count", odd_num_count)
```

## 3. Separate the numbers and strings in the list [10, 20, 'python', '10000', 'program', 300] and save them in two different lists.

```python
type('python')
```

```python
input_list =  [10, 20, 'python', '10000', 'program', 300]
num_list = []
string_list = []

for item in input_list:
  if type(item) == str:
    string_list.append(item)
  else:
    num_list.append(item)

print("String in list are ", string_list)
print("Numbers in list are ", num_list)
```

## 4. Marks of students in class are stored in the list [22, 65, 40, 99, 100, 50, 78, 35].

###   a) Write a Program to print the index of the items in that list that are greater than 35.

```python
marks_list =  [22, 65, 40, 99, 100, 50, 78, 35]
for i in range(0, len(marks_list)):
  item = marks_list[i]
  if item > 35:
    print(i)
```

###   b) Write a Program to store the marks that are greater than 35 in one list and less than or equal to 35 in another list

```python
marks_list =  [22, 65, 40, 99, 100, 50, 78, 35]
list_greater = []
list_lesser = []
for item in marks_list:
  if item > 35:
    list_greater.append(item)
  else:
    list_lesser.append(item)
print("Marks greater than 35", list_greater)
print("Marks lessthan or equal to 35", list_lesser)


```
